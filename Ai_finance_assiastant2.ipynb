```python
import fitz  # PyMuPDF
import os
import re
import torch
import openai
import open_clip
import chromadb
import llmgraph
from google.colab import files
from PIL import Image
from transformers import CLIPTokenizer # Import CLIPTokenizer

# Set OpenAI API Key
openai.api_key = ""
# Load CLIP Model and Tokenizer
device = "cuda" if torch.cuda.is_available() else "cpu"
model, _, preprocess = open_clip.create_model_and_transforms("ViT-B/32", pretrained="openai")
model = model.to(device)
#model, _, preprocess = open_clip.create_model_and_transforms("ViT-B-32", pretrained="openai") # Changed model name to ViT-B-32. This is causing the error
# Initialize the CLIP tokenizer
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")
# Initialize ChromaDB
chroma_client = chromadb.PersistentClient(path="chroma_db")
text_collection = chroma_client.get_or_create_collection(name="text_embeddings")
image_collection = chroma_client.get_or_create_collection(name="image_embeddings")

# AI Finance Assistant Prompt
finance_prompt = """
You are an advanced AI finance assistant, skilled in financial analysis, investment strategies, budgeting, risk assessment, and market trend analysis. Your goal is to provide accurate, insightful, and well-structured financial advice based on user queries. You adhere to real-time data where possible, follow financial best practices, and communicate in a professional yet easy-to-understand manner.

Capabilities & Guidelines:
1. Investment & Market Analysis
2. Personal Finance & Budgeting
3. Risk Management & Portfolio Optimization
4. Tax & Retirement Planning
5. Live Market Integration (if connected to APIs)
6. Clear & Actionable Responses
7. Ethical & Legal Compliance

Response Format:
📌 Summary
📈 Analysis
💡 Recommendations
📊 Data & Visuals
⚠️ Risk Considerations
"""

# Extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text_chunks = []
    for page_num in range(len(doc)):
        text = doc[page_num].get_text("text").strip()
        if text:
            sentences = re.split(r'(?<=[.!?]) +', text)
            chunks = [" ".join(sentences[i:i+3]) for i in range(0, len(sentences), 3)]
            for chunk in chunks:
                text_chunks.append((chunk, page_num + 1))
    return text_chunks

# Extract images from PDF
def extract_images_from_pdf(pdf_path, output_folder="extracted_images"):
    doc = fitz.open(pdf_path)
    os.makedirs(output_folder, exist_ok=True)
    image_paths = []
    for page_num in range(len(doc)):
        for img_index, img in enumerate(doc[page_num].get_images(full=True)):
            xref = img[0]
            base_image = doc.extract_image(xref)
            img_filename = f"{output_folder}/page_{page_num+1}_img_{img_index+1}.{base_image['ext']}"
            with open(img_filename, "wb") as img_file:
                img_file.write(base_image["image"])
            image_paths.append((img_filename, page_num + 1))
    return image_paths

def get_text_embeddings(text_chunks, pdf_name):
    for i, (chunk, page_num) in enumerate(text_chunks):
        # Truncate the text to the maximum sequence length
        tokens = tokenizer(chunk, truncation=True, max_length=77, return_tensors="pt", padding="max_length")  # Add padding
        text_tokens = tokens.input_ids.to(device)

        with torch.no_grad():
            embedding = model.encode_text(text_tokens).cpu().numpy().tolist()[0]
        text_collection.add(
            ids=[f"text_{pdf_name}_{i}"],
            embeddings=[embedding],
            metadatas=[{"text": chunk, "source": pdf_name, "page": page_num}]
        )

# Generate image embeddings using CLIP
def get_image_embeddings(image_paths, pdf_name):
    for i, (img_path, page_num) in enumerate(image_paths):
        image = preprocess(Image.open(img_path)).unsqueeze(0).to(device)
        with torch.no_grad():
            embedding = model.encode_image(image).cpu().numpy().tolist()[0]
        image_collection.add(
            ids=[f"image_{pdf_name}_{i}"],
            embeddings=[embedding],
            metadatas=[{"image_path": img_path, "source": pdf_name, "page": page_num}]
        )

def retrieve_similar_content(query, top_k=5):
    tokens = tokenizer(query, truncation=True, max_length=77, return_tensors="pt", padding="max_length")
    query_tokens = tokens['input_ids'].to(device) # Access the 'input_ids' to get the token IDs and move to the device
    with torch.no_grad():
        query_embedding = model.encode_text(query_tokens).cpu().numpy().tolist()[0]
    text_results = text_collection.query(query_embeddings=[query_embedding], n_results=top_k)
    image_results = image_collection.query(query_embeddings=[query_embedding], n_results=top_k)
    return text_results, image_results

# Create a knowledge graph
def create_knowledge_graph(query, text_results, image_results):
    graph = llmgraph.Graph()
    query_node = graph.add_node(query, node_type="Query")
    for meta in text_results.get("metadatas", [[]])[0]:
        if "text" in meta:
            text_node = graph.add_node(meta["text"], node_type="Text Chunk")
            graph.add_edge(query_node, text_node, relationship="Relevant to")
    for meta in image_results.get("metadatas", [[]])[0]:
        if "image_path" in meta:
            image_node = graph.add_node(meta["image_path"], node_type="Image")
            graph.add_edge(query_node, image_node, relationship="Relevant to")
    return graph.visualize()

def generate_contextual_response(query, text_results, image_results):
    context_text = "\n".join([meta["text"] for meta in text_results.get("metadatas", [[]])[0] if "text" in meta])
    context_images = ", ".join([meta["image_path"] for meta in image_results.get("metadatas", [[]])[0] if "image_path" in meta])
    prompt = f"""
    {finance_prompt}
    Query: {query}

    Text Context:
    {context_text}

    Image Context (paths to relevant images):
    {context_images if context_images else "No relevant images found."}

    Provide a well-structured response based on the given context.
    """
    # Updated OpenAI API call for version 1.0.0 or later:
    response = openai.ChatCompletion.create( # Use openai.ChatCompletion instead of openai.Completion
        model="gpt-3.5-turbo", # Use gpt-3.5-turbo for chat completion in openai>=1.0.0
        messages=[{"role": "user", "content": prompt}], # Provide the prompt as a message
        max_tokens=150  # Adjust as needed
    )
    return response.choices[0].message.content # Access the response content from the message object


# Upload and process PDF
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]
pdf_name = os.path.splitext(pdf_path)[0]

text_chunks = extract_text_from_pdf(pdf_path)
image_files = extract_images_from_pdf(pdf_path)

get_text_embeddings(text_chunks, pdf_name)
get_image_embeddings(image_files, pdf_name)

print("Processing complete. Data stored in ChromaDB.")
```

```python
def query_and_display():
    print("You can now query the system. Type 'exit' to quit.")
    while True:
        query = input("Enter your query: ")
        if query.lower() == 'exit':
            break
        text_results, image_results = retrieve_similar_content(query)
        response = generate_contextual_response(query, text_results, image_results)

        print("\nAI Response:")
        print(response)

        print("\nRelevant Text Chunks:")
        for i, meta in enumerate(text_results["metadatas"][0], 1):
            if "text" in meta:
                print(f"{i}. {meta['text']}")

        print("\nRelevant Images:")
        for meta in image_results["metadatas"][0]:
            if "image_path" in meta:
                display(Image.open(meta["image_path"]))
```
